```python
df_cleaned = df_tracks.drop(['key', 'mode', 'explicit'], axis=1)

numeric_df = df_cleaned.select_dtypes(include=['number'])

# Step 3: Calculate the correlation matrix
heatmap_df = numeric_df.corr(method='pearson')
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(14,6))
heatmap= sns.heatmap(heatmap_df,fmt =".1g",annot=True, vmin=-1,vmax=1,center = 0, cmap ='inferno',linewidths=1, linecolor = 'Black')
heatmap.set_title("Correlation Heatmap between Variables")

plt.show()
```

```python


def convert_date(date_str):
    try:
        # Try the detailed date format first
        return pd.to_datetime(date_str, format='%Y-%m-%d', errors='raise')
    except ValueError:
        try:
            # If that fails, try just the year format
            return pd.to_datetime(date_str, format='%Y', errors='raise')
        except ValueError:
            # Handle other formats or return NaT for not-a-time
            return pd.NaT

```

```python
df_tracks['dates']= df_tracks['release_date'].apply(convert_date)
```

```python
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
pip show seaborn
```

```python
hist_plot = sns.displot(years,discrete=True, aspect=2, height=5,kind="hist").set(title="Songs per year")


ax = hist_plot.ax  # Get the Axes object
n, bins, patches = ax.hist(years, bins=len(years.unique()), align='left', rwidth=0.8)

# Find the bin with the highest count
max_bin_index = n.argmax()
max_bin_value = n[max_bin_index]
max_bin_center = (bins[max_bin_index] + bins[max_bin_index + 1]) / 2

# Annotate the highest bin
ax.annotate(f'{int(max_bin_value)}',
            xy=(max_bin_center, max_bin_value),
            xytext=(0, 5),  # Offset label above the bar
            textcoords='offset points',
            ha='center', va='bottom',
            fontsize=12, color='black')

plt.show()
```

```python
total_dr = df_tracks.duration
fig_dim = (18,7)
fig,ax = plt.subplots(figsize = fig_dim)
fig = sns.barplot(x= years, y = total_dr, ax = ax,err_kws={'linewidth': 0}).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style ='whitegrid')
fig_dim = (18,7)
fig,ax = plt.subplots(figsize = fig_dim)
fig = sns.lineplot(x= years, y = total_dr, ax = ax).set(title="Year vs Duration")
plt.xticks(rotation=90)
```
